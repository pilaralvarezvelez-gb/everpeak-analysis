## Ejercicio 3 · Crear una copia limpia \(15 min\)

Regla de trabajo: no modificamos el dataset original directamente. Creamos `df_limpio` para preservar la trazabilidad.

Aplicaremos cuatro decisiones:

1. Convertir `fecha` a tipo fecha.
2. Normalizar textos en `canal`.
3. Completar `metodo_pago` con `No informado`.
4. Completar variables numéricas con la mediana.

````python
# TODO estudiante: escribe aquí tu código.
# Sugerencia:
# 1. Lee la consigna anterior.
# 2. Ejecuta paso a paso.
# 3. Valida el resultado antes de interpretarlo.

df_limpio = df.copy()
df_limpio.info()
````

````python
# TODO estudiante: escribe aquí tu código.
# Sugerencia:
# 1. Lee la consigna anterior.
# 2. Ejecuta paso a paso.
# 3. Valida el resultado antes de interpretarlo.

df_limpio = df.copy()

df_limpio["fecha"] = pd.to_datetime(
    df_limpio["fecha"],
    errors ="coerce",
    utc=True
    )
df_limpio["fecha"].dtype
````

````python
df_limpio['canal']= df_limpio['canal'].str.replace('web','Web')
df_limpio['canal']= df_limpio['canal'].str.replace('tienda física','Tienda física')
df_limpio['canal']= df_limpio['canal'].str.replace('App ', 'App')
df_limpio['canal'].unique()
````

````python
df_limpio['metodo_pago']= df_limpio['metodo_pago'].fillna('No informado')
df_limpio['metodo_pago'].unique()
````

````python
df_limpio.info()
````

````python
df_limpio['edad_cliente'] = df_limpio['edad_cliente'].fillna(df_limpio['edad_cliente'].median())
df_limpio['satisfaccion'] = df_limpio['satisfaccion'].fillna(df_limpio['satisfaccion'].median())
df_limpio['dias_entrega'] = df_limpio['dias_entrega'].fillna(df_limpio['dias_entrega'].median())
````

````python
medianas = df_limpio.select_dtypes(include='number').median()
print(medianas)
````

## Ejercicio 4 · Automatización con `for`, `while` y funciones \(10 min\)

En análisis de datos hay tareas repetitivas: revisar varias columnas, calcular varios resúmenes o generar varios gráficos. Para eso usamos automatización básica.

Primero recorreremos columnas numéricas con `for`.

columnas con operaciones repetitivas que se pueden automatizar: ingreso, descuento, margen

````python
numeric_cols = df_limpio.select_dtypes(include=['number']).columns

print("Columnas numéricas en df_limpio:")
for col in numeric_cols:
    print(col)
````

````python
print('\nEstadísticas descriptivas para columnas numéricas:')
for col in numeric_cols:
    print(f"\nColumna: {col}")
    print(f"  Media: {df_limpio[col].mean():,.2f}")
    print(f"  Mediana: {df_limpio[col].median():,.2f}")
    print(f"  Mínimo: {df_limpio[col].min():,.2f}")
    print(f"  Máximo: {df_limpio[col].max():,.2f}")
````

````python
# Definimos una lista de las columnas que queremos procesar con el bucle while
columns_to_process_while = ['ingreso', 'descuento_pct', 'margen']

print("\nProcesando columnas con bucle while:")

# Mientras la lista de columnas no esté vacía
while columns_to_process_while:
    # Tomamos la primera columna de la lista
    current_col = columns_to_process_while.pop(0) # .pop(0) remueve y devuelve el primer elemento

    # Realizamos alguna operación, por ejemplo, imprimir el tipo de dato y la media
    if current_col in df_limpio.columns:
        print(f"  Columna '{current_col}':")
        print(f"    Tipo de dato: {df_limpio[current_col].dtype}")
        print(f"    Valor medio: {df_limpio[current_col].mean():,.2f}")
    else:
        print(f"  Columna '{current_col}' no encontrada en el DataFrame.")

print("\nBucle while completado. Todas las columnas han sido procesadas.")
````
